# Music Stats

General stats of my music library (this is mostly just a Jupyter Notebook exercise)

## I. Import libraries, initialize variables

```python
import os
import re
from pandas import DataFrame
from tinytag import TinyTag

MUSIC_DIRECTORY = u'~/btsync/music'
```

## II. Read and load music

#### A. List files

```python
music_path = os.path.expanduser(MUSIC_DIRECTORY)

music_list = []
for root, dirs, files in os.walk(music_path):
    for f in files:
        if re.match(r'.*(mp3|wav|ogg|flac|m4a)$', f, re.IGNORECASE):
            music_list.append(os.path.join(root, f))
            
print "Files: %s" % len(music_list)
```

#### B. Extract and load file tags
Extract tags using TinyTag, then load them on to a DataFrame

```python
def fetch_tags(filepath):
    tag = TinyTag.get(filepath)
    return {
        'album': tag.album,
        'albumartist': tag.albumartist,
        'artist': tag.artist,
        'audio_offset': tag.audio_offset,
        'bitrate': tag.bitrate,
        'disc': tag.disc,
        'disc_total': tag.disc_total,
        'duration': tag.duration,
        'filesize': tag.filesize,
        'genre': tag.genre,
        'samplerate': tag.samplerate,
        'title': tag.title,
        'track': tag.track,
        'track_total': tag.track_total,
        'year': tag.year,
    }

music_df = DataFrame(data=map(fetch_tags, music_list))

# Check DataFrame contents
music_df
```
